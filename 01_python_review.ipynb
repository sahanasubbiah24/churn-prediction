```python
# List comprehension approach

squares_comprehension = [number ** 2 for number in range(1, 11)]

print(squares_comprehension)
```

```python
# Dictionary comprehension approach

square_dictionary = {
    number: number ** 2
    for number in range(1, 6)
}

print(square_dictionary)
```

```python
customer_names = ["Alice", "Benjamin", "Cara", "Daniel"]

name_lengths = {
    name: len(name)
    for name in customer_names
}

print(name_lengths)
```

```python
customer_spending = {
    "Alice": 120,
    "Bob": 45,
    "Cara": 210,
    "Daniel": 75
}

high_value_customers = {
    name: spending
    for name, spending in customer_spending.items()
    if spending >= 100
}

print(high_value_customers)
```

```python
square = lambda number: number ** 2

print(square(7))
```

```python
calculate_total = lambda price, quantity: price * quantity

print(calculate_total(19.99, 3))
```

```python
numbers = [1, 2, 3, 4, 5]

squared_numbers = list(
    map(lambda number: number ** 2, numbers)
)

print(squared_numbers)
```

```python
numbers = [1, 2, 3, 4, 5, 6, 7, 8]

even_numbers = list(
    filter(lambda number: number % 2 == 0, numbers)
)

print(even_numbers)
```

```python
customers = ["Alice", "Bob", "Cara", "Daniel"]

for index, customer in enumerate(customers):
    print(index, customer)
```

```python
for position, customer in enumerate(customers, start=1):
    print(f"{position}. {customer}")
```

```python
customer_names = ["Alice", "Bob", "Cara"]
monthly_charges = [29.99, 54.50, 72.25]

for name, charge in zip(customer_names, monthly_charges):
    print(f"{name} pays ${charge:.2f} per month.")
```

```python
customer_charge_dictionary = dict(
    zip(customer_names, monthly_charges)
)

print(customer_charge_dictionary)
```

```python
customer_name = "Alice"
monthly_charge = 54.50
tenure_months = 18

message = (
    f"{customer_name} has been a customer for "
    f"{tenure_months} months and pays "
    f"${monthly_charge:.2f} per month."
)

print(message)
```

```python
print(
    f"Estimated annual charge: "
    f"${monthly_charge * 12:.2f}"
)
```

```python
def calculate_annual_charge(monthly_charge):
    """
    Calculate a customer's estimated annual charge.

    Parameters:
        monthly_charge: Customer's monthly charge.

    Returns:
        Estimated annual charge.
    """
    return monthly_charge * 12


annual_charge = calculate_annual_charge(59.99)

print(f"${annual_charge:.2f}")
```

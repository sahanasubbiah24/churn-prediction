```python
import pandas as pd
```

```python
customer_data = {
    "customer_id": [
        "C001",
        "C002",
        "C003",
        "C004",
        "C005",
        "C006",
        "C007",
        "C008"
    ],
    "gender": [
        "Female",
        "Male",
        "Female",
        "Male",
        "Female",
        "Male",
        "Female",
        "Male"
    ],
    "tenure_months": [
        2,
        24,
        13,
        48,
        6,
        36,
        1,
        60
    ],
    "monthly_charge": [
        29.99,
        75.50,
        54.25,
        89.99,
        45.00,
        66.75,
        20.00,
        99.50
    ],
    "contract_type": [
        "Month-to-month",
        "One year",
        "Month-to-month",
        "Two year",
        "Month-to-month",
        "One year",
        "Month-to-month",
        "Two year"
    ],
    "churn": [
        "Yes",
        "No",
        "Yes",
        "No",
        "Yes",
        "No",
        "Yes",
        "No"
    ]
}

customers = pd.DataFrame(customer_data)

customers
```

```python
high_charge_customers = customers[
    customers["monthly_charge"] > 70
]

high_charge_customers
```

```python
churned_customers = customers[
    customers["churn"] == "Yes"
]

churned_customers
```

```python
new_high_charge_customers = customers[
    (customers["tenure_months"] < 12) &
    (customers["monthly_charge"] > 40)
]

new_high_charge_customers
```

```python
customers["estimated_annual_charge"] = (
    customers["monthly_charge"] * 12
)

customers
```

```python
customers["estimated_annual_charge"] = (
    customers["estimated_annual_charge"].round(2)
)

customers
```

```python
def categorize_tenure(tenure_months):
    if tenure_months < 12:
        return "New"
    elif tenure_months < 36:
        return "Established"
    else:
        return "Long-term"
```

```python
customers["tenure_group"] = customers[
    "tenure_months"
].apply(categorize_tenure)

customers
```

```python
customers["churn_numeric"] = customers[
    "churn"
].map({
    "No": 0,
    "Yes": 1
})

customers
```

```python
customers_sorted = customers.sort_values(
    by="monthly_charge",
    ascending=False
)

customers_sorted
```

```python
customers.sort_values(
    by=["contract_type", "tenure_months"],
    ascending=[True, False]
)
```

```python
average_charge_by_churn = customers.groupby(
    "churn"
)["monthly_charge"].mean()

average_charge_by_churn
```

```python
contract_summary = customers.groupby(
    "contract_type"
).agg(
    customer_count=("customer_id", "count"),
    average_tenure=("tenure_months", "mean"),
    average_monthly_charge=("monthly_charge", "mean"),
    churn_rate=("churn_numeric", "mean")
)

contract_summary
```

```python
contract_summary["churn_rate_percent"] = (
    contract_summary["churn_rate"] * 100
).round(1)

contract_summary
```

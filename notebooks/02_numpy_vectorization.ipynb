```python
import numpy as np
```

```python
monthly_charges = np.array([
    29.99,
    54.50,
    72.25,
    89.99,
    45.00
])

print(monthly_charges)
print(type(monthly_charges))
```

```python
print("Shape:", monthly_charges.shape)
print("Dimensions:", monthly_charges.ndim)
print("Number of values:", monthly_charges.size)
print("Data type:", monthly_charges.dtype)
```

```python
annual_charges = monthly_charges * 12

print(annual_charges)
```

```python
discounted_charges = monthly_charges * 0.90
increased_charges = monthly_charges * 1.05

print("10% discount:", discounted_charges)
print("5% increase:", increased_charges)
```

```python
rounded_annual_charges = np.round(
    annual_charges,
    decimals=2
)

print(rounded_annual_charges)
```

```python
monthly_charges_list = [
    29.99,
    54.50,
    72.25,
    89.99,
    45.00
]

annual_charges_loop = []

for charge in monthly_charges_list:
    annual_charges_loop.append(charge * 12)

print(annual_charges_loop)
```

```python
monthly_charges_array = np.array(monthly_charges_list)

annual_charges_numpy = monthly_charges_array * 12

print(annual_charges_numpy)
```

```python
monthly_charges = np.array([
    29.99,
    54.50,
    72.25,
    89.99,
    45.00
])

high_charge_mask = monthly_charges > 60

print(high_charge_mask)
```

```python
print(monthly_charges[monthly_charges > 60])
```

```python
medium_charges = monthly_charges[
    (monthly_charges >= 40) &
    (monthly_charges <= 75)
]

print(medium_charges)
```

```python
print("Minimum:", np.min(monthly_charges))
print("Maximum:", np.max(monthly_charges))
print("Average:", np.mean(monthly_charges))
print("Median:", np.median(monthly_charges))
print("Total:", np.sum(monthly_charges))
print("Standard deviation:", np.std(monthly_charges))
```

```python
average_charge = np.mean(monthly_charges)

print(f"Average monthly charge: ${average_charge:.2f}")
```

```python
customer_metrics = np.array([
    [12, 29.99],
    [24, 54.50],
    [36, 72.25],
    [6, 89.99]
])

print(customer_metrics)
```
